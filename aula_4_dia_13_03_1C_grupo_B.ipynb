```python
# escrever um programa que faça cálculos aritméticos

# entrada

a= int (input("digite o primeiro número"))
b= int (input("digite o segundo número"))

# processamento
soma= a + b
subtracao= a - b
multiplicacao= a * b
divisao= a / b

# saída

print("---------------CALCULADORA SIMPLES---------------")
print("")
print ("---------------soma---------------")
print("")
print (f"o resultado da soma de {a} mais {b} é igual a:{soma}")
print("")
print("---------------subtração---------------")
print("")
print(f"o resultado da subtração de {a} menos {b} é igual: {subtracao}")
print("")
print("---------------multiplicação---------------")
print("")
print (f"o resultado da multiplicação de {a} vezes {b} é igual:{multiplicacao}")
print("")
print("---------------divisão---------------")
print("")
print(f"o resultado da divisão de {a} dividido por {b} é igual: {divisao}")
```
