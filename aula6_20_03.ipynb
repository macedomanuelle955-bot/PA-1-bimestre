#Média de 3 números
#Entrada
ler 3 números

#Processamento
Calcular a média dos 3

#Saída
Mostrar o resultado do cálculo da média

```python
#Entrada
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
num3 = int(input("Digite o terceiro número: "))

#Processamento
resultado = (num1 + num2 + num3) / 3

#Saída
print(f"O resultado da média é: {resultado}")
```

#Divisão inteira e resto

#Entrada
Ler os dois números inteiros

#Processamento
Dividir dois números inteiros  

#Saída
Mostrar o resultado da divisão

```python
#Entrada
numero = int(input("Digite um número: "))

#Processamento
dobro = numero * 2
metade = numero / 2
resto = numero % 3

#Saída
print("Dobro:", dobro)
print("Metade:", metade)
print("Resto da divisão por 3:", resto)
```

#Consumo de combustível

```python
#Entrada
distancia = float(input("Digite a distância (km): "))
litros = float(input("Digite o combustível gasto (litros): "))

#Processamento
if litros != 0:
    consumo = distancia / litros
    #Saída
    print("Consumo médio (km/l):", consumo)
else:
    #Saída para caso de litros ser 0
    print("Não é possível calcular o consumo com zero litros de combustível.")
```
