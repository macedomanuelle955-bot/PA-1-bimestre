```python
#Contagem regressiva, o usuario digita um número e o programa imprimi do número até o zero

numero= int(input("digite um numero inteiro positivo:")
)
print(f"contagem regressiva de{numero} até 0:")

for i in range (numero, -1,-1):
  print(i)
```

#Questão 5

```python
#Soma,o usuario digita um número e o programa soma os números do zero até o número

numero = int(input("Digite um número: "))

soma = 0
for i in range(numero + 1):
    soma += i

print(f"A soma dos números de 0 até {numero} é: {soma}")
```

#Questão 6

```python
#Tabuada, o usuário digita um número e o programa imprimi a tabuada do número

a = int(input("Digite um número para ver a tabuada: "))

for i in range(1, 11):
    print(f"{a} x {i} = {a * i}")
```

#Questão 8

```python
#Contar vogais, o usuário digita uma palavra e o programa conta as vogais

palavra = input("Digite uma palavra: ")
vogais = "aeiou"
contador_vogais = 0

for letra in palavra:
    if letra in vogais:
        contador_vogais += 1
        print(f"a palavra  '{palavra}' tem {contador_vogais}vogais")
```

#Questão 9

```python
#Maior número digitado, o usuário digita quatro números e o programa imprimi o maior número digitado
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
num3 = int(input("Digite o terceiro número: "))
num4 = int(input("Digite o quarto número: "))

maior = max(num1, num2, num3, num4)

print(f"O maior número digitado foi o: {maior}")
```

#Questão 10

```python
#O usuário digita um número entre 01 e 100 e o programa mostra quais são divisíveis por 3 e 5
numero_limite = int(input("Digite um número entre 1 e 100: "))

if 1 <= numero_limite <= 100:
    print(f"Números divisíveis por 3 e 5 até {numero_limite}:")
    for i in range(1, numero_limite + 1):
        if i % 3 == 0 and i % 5 == 0:
            print(i)
else:
    print("Por favor, digite um número entre 1 e 100.")
```
