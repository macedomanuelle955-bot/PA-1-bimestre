```python
#Contagem regressiva, o usuario digita um número e o programa imprimi do número até o zero

numero= int(input("digite um numero inteiro positivo:")
)
print(f"contagem regressiva de{numero} até 0:")

for i in range (numero, -1,-1):
  print(i)
```

#Questão 3

```python
#Números pares, o usuario digita um número e o programa imprimi os números pares
numero= int(input("digite um número inteiro positivo:"))
print(f"Selecionando números pares até {numero}:")

for i in range (numero + 1):
  if i % 2 ==0:
    print(i)
```
