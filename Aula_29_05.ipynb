```python
import random

numero_secreto= random.randint(1, 10)

palpite= int(input("Digite um palpite:"))

if palpite == numero_secreto:
 print("Você acertou!")

else:
  print("Você errou!")

if palpite > numero_secreto:
  print("Você acertou!")
```
